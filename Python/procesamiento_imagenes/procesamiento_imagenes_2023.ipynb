```python
import cv2
import matplotlib.pyplot as plt 


#imagen en BGR
img = cv2.imread('lena.png')

hist_blue = cv2.calcHist([img], [0], None, [256], [0,256])
hist_green = cv2.calcHist([img], [1], None, [256], [0,256])
hist_red = cv2.calcHist([img], [2], None, [256], [0,256])

#grafico
plt.figure(dpi=150)
plt.plot(hist_blue, color = 'blue', label='blue')
plt.plot(hist_green, color = 'green',label='green')
plt.plot(hist_red, color = 'red', label='red')
plt.legend()
plt.grid(alpha=0.2)
plt.xlim(0,255)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#funcion gamma (implementada por nosotros)
def gamma(img, g):
    norm = img/255
    out = norm**g
    #ojo que retornamos valores decimales entre 0 y 1
    return out


img = cv2.imread('cameraman.png')
out = gamma(img, 0.4)

plt.imshow(out)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#lectura de la imagen
img = cv2.imread('lena.png')
m,n,ch = img.shape

blue, green, red= cv2.split(img)

eq_blue  = cv2.equalizeHist(blue)
eq_green = cv2.equalizeHist(green)
eq_red   = cv2.equalizeHist(red)

#seteamos el output para tres canales de tipo uint8
out = cv2.merge([eq_red, eq_green, eq_blue])

#resultados
fig, ax = plt.subplots(nrows=1, ncols=2, dpi=100)
fig.set_size_inches(10,15)
ax[0].imshow(img[:,:,[2,1,0]])
ax[0].set_title("Original")
ax[1].imshow(out)
ax[1].set_title("Ecualizada")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('cameraman.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#normalizamos la imagen original entre 0 y 1
X = cv2.normalize(gray.astype('float'), None, 0, 1, cv2.NORM_MINMAX)



A = cv2.blur(X**2, (5,5))
B = cv2.blur(X, (5,5))**2

res = np.sqrt(A-B)

#volvemos a normalizar para mejorar el rango de salida entre 0 y 1
res = cv2.normalize(res, None, 0, 1, cv2.NORM_MINMAX)

plt.figure(figsize=(10,8))
plt.imshow(res, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#lectura de la imagen
img = cv2.imread('cameraman.png', cv2.IMREAD_GRAYSCALE)
m,n = img.shape

#incorporamos ruido a la imagen original
mat_noise=np.random.rand(m,n)  #Agregamos ruido aleatorio entre 0 y 1
sp_noise_white= np.uint8(np.where(mat_noise>=0.8, 255,0))
sp_noise_black= np.uint8(np.where(mat_noise>=0.1,   1,0))
noise_img    = cv2.multiply(img,sp_noise_black)
noise_img_sp = cv2.add(noise_img,sp_noise_white)


#aplicamos el filtro
img_median   = cv2.medianBlur(noise_img_sp,(5))

fig, ax= plt.subplots(nrows=1, ncols=3)
fig.set_size_inches(15,20)
ax[0].imshow(img, cmap='gray')
ax[0].set_title('imagen original')
ax[1].imshow(noise_img_sp, cmap='gray')
ax[1].set_title('imagen con ruido sal y pimienta')
ax[2].imshow(img_median, cmap="gray")
ax[2].set_title('imagen filtrada')
plt.show()
```
