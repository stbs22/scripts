```python
import numpy as np
import matplotlib.pyplot as plt
tiempo = np.linspace(0,1, num=1000)
funcion = 0.5*np.sin(2*np.pi*tiempo*25)

plt.plot(tiempo, funcion)
plt.show()
```

```python
F = np.fft.fft(funcion)
N = len(F)
p = np.linspace(0, 49, dtype=int, num=50)
plt.bar(x=p, height=abs(F[p])/N*2)
plt.show()
```

```python
R = np.random.random(1000)
funtion = 0.5*np.sin(2*np.pi*25*tiempo)+np.sin(2*np.pi*15*tiempo)+0.9*R
plt.plot(tiempo, funtion)
plt.show()
```

```python
F = np.fft.fft(funcion)
N = len(F)
p = np.linspace(0, 49, dtype=int, num=50)
plt.bar(x=p, height=abs(F[p])/N*2)
plt.show()
```
